```python
import numpy as np

# Generate two sets of data with points close to -0.01 and +0.01
data1 = np.random.rand(50, 2) * 0.1 + np.array([-0.01, -0.01]) # Data around -0.01
data2 = np.random.rand(50, 2) * 0.1 + np.array([0.04, 0.04]) # Data around +0.04
```

```python
import matplotlib.pyplot as plt

# Plot the new data with different colors
plt.scatter(data1[:, 0], data1[:, 1], color='blue', label='Data Set 1')
plt.scatter(data2[:, 0], data2[:, 1], color='red', label='Data Set 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('New Separated Data Sets (Close Points)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import numpy as np
import matplotlib.pyplot as plt

# Combine the new data and create labels
X = np.vstack((data1, data2))
y = np.hstack((np.zeros(len(data1)), np.ones(len(data2))))

# Split data into training and testing sets (optional, but good practice)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Retrain Logistic Regression model
model_lr_new_data = LogisticRegression()
model_lr_new_data.fit(X_train, y_train)

# Retrain MLPClassifier with ReLU model
model_mlp_relu_new_data = MLPClassifier(hidden_layer_sizes=(10,), activation='relu', max_iter=1000, random_state=42)
model_mlp_relu_new_data.fit(X_train, y_train)

# Evaluate models on the new test data
y_pred_lr_new_data = model_lr_new_data.predict(X_test)
accuracy_lr_new_data = accuracy_score(y_test, y_pred_lr_new_data)
print(f"Model Accuracy (Logistic Regression on New Data): {accuracy_lr_new_data:.2f}")

y_pred_relu_mlp_new_data = model_mlp_relu_new_data.predict(X_test)
accuracy_relu_mlp_new_data = accuracy_score(y_test, y_pred_relu_mlp_new_data)
print(f"Model Accuracy (MLPClassifier with ReLU on New Data): {accuracy_relu_mlp_new_data:.2f}")
```

```python
# Plot the decision boundary of the Logistic Regression model on new data
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5 # Adjusted scale
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5 # Adjusted scale
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z_lr_new = model_lr_new_data.predict(np.c_[xx.ravel(), yy.ravel()])
Z_lr_new = Z_lr_new.reshape(xx.shape)

plt.contourf(xx, yy, Z_lr_new, cmap='coolwarm', alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=20, edgecolors='k')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Logistic Regression Decision Boundary (New Data)')
plt.show()
```

```python
# Plot the decision boundary of the MLPClassifier with ReLU on new data
x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5 # Adjusted scale
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5 # Adjusted scale
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))

Z_relu_new = model_mlp_relu_new_data.predict(np.c_[xx.ravel(), yy.ravel()])
Z_relu_new = Z_relu_new.reshape(xx.shape)

plt.contourf(xx, yy, Z_relu_new, cmap='coolwarm', alpha=0.8)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm', s=20, edgecolors='k')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('MLPClassifier with ReLU Decision Boundary (New Data)')
plt.show()
```
